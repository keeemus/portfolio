## 데이터 설명
### 건강보험공단에서 매년 제공하는 건강보험가입자의 검진데이터
0. 기준년도 : 제공된 데이터 측정년도
1. 가입자 일련번호 : 건강보험에 가입된 가입자에 부여된 번호
2. 시도코드 : 수검자의 거주지
3. 성별코드 : 수검자의 성별
4. 연령대 코드 : 기준년도에 수검자의 나이를 5세 단위로 범주화하여 구분한 코드(5세 단위)
5. 신장 : 수검자의 신장(5cm 단위)
6. 체중 : 수검자의 체중(5kg 단위)
7. 허리둘레 : 수검자의 허리둘레
8. 시력(좌) : 수검자의 왼쪽 눈 시력
9. 시력(우) : 수검자의 오른쪽 눈 시력
10. 청력(좌) : 수검자의 왼쪽 귀 청력
11. 청력(우) : 수검자의 오른쪽 귀 청력
12. 수축기 혈압 : 수검자의 최고 혈압으로 심장이 수축해서 강한 힘으로 혈액을 동맥에 보낼 때 혈관의 내압
13. 이완기 혈압 : 수검자의 최저 혈압으로 수축기 혈압과 반대
14. 식전혈당(공복혈당) : 검진자 식사 전 혈당(혈액 100ml당 함유되어있는 포도당의 농도) 수치
15. 총 콜레스테롤 : 혈청 중의 함량된 전체 콜레스테롤
16. 트리글리세라이드 : 단순지질(중성지질)
17. HDL 콜레스테롤 : 고밀도 단백질에 포함되는 콜레스테롤, 나쁜 콜레스테롤을 없애는 역할
18. LDL 콜레스테롤 : 저밀도 단백질에 포함되는 콜레스테롤, 과다할 경우 혈관벽에 쌓여 각종 질병을 야기하는 성분
19. 혈색소 : 혈액이나 혈구 속에 존재하는 색소단백질, 혈중 산소운반체 역할
20. 요단백 : 소변에 단백질이 섞여 나오는 현상
21. 혈청크레아티닌 : 근육의 발육과 운동에 관여하는 성분, 과도할 경우 신장기능장애 유발
22. (혈청지오티)AST : 간 기능을 나타내는 혈액검사상의 수치, 간세포 및 심장, 신장, 뇌, 근육 등에 존재하며 세포들이 손상 받는 경우 농도가 증가함
23. (혈청지오티)ALT : 간 기능을 나타내는 혈액검사상의 수치, 간세포에 주로 존재하는 효소이고 세포들이 손상 받는 경우 농도가 증가함
24. 감마 지티피 : 간 기능을 나타내는 혈액검사상의 수치, 간세포 장애 발생시 농도가 증가함.
25. 흡연상태 : 수검자의 흡연여부를 나타냄
26. 음주여부 : 수검자의 음주여부를 나타냄
27. 구강검진 수검여부 : 수검자의 구강검진 선택여부
28. 치아우식증유무 : 수검자의 치아우식증 유무
29. 치석 : 수검자의 치석 여부
30. 데이터 공개일자 : 데이터 작성 기준일자

## 1. 예측인자가 아닌 특성 제거

```python
del_column = ['기준년도', '가입자 일련번호', '시도코드', '체중(5Kg 단위)', '시력(좌)', '시력(우)', '청력(좌)', '청력(우)', '혈색소', '요단백'
              ,'혈청크레아티닌', '(혈청지오티)AST', '(혈청지오티)ALT', '감마 지티피', '구강검진 수검여부', '치아우식증유무', '치석', '데이터 공개일자']
```

### 결측치 특성 제거

결측치가 60% 넘어가는 특성 제거</br>
df_deleted : 결측치 특성 제거한 데이터셋</br>
df : 결측치 특성 제거하지 않은 데이터셋</br>

```python
df_deleted = df.drop(['총 콜레스테롤', '중성지방', 'HDL 콜레스테롤', 'LDL 콜레스테롤'], axis=1)
```

```python
# 이외의 결측치는 평균값보다 중앙값으로 대체하는 것이 이상치에 덜 민감하기 때문에 각 컬럼의 중앙값으로 대체


df_deleted = df_deleted.fillna(df_deleted.median())
df_deleted
```

### 결측치 특성 대체

특성의 결측치 비율이 50%가 넘어갈 경우 특정값으로 결측치를 대체하는 것이 데이터 분포에 큰 영향을 미치기때문에 특성 자체를 삭제하는 것이 현명한 방법</br>
하지만 중성지방, 콜레스테롤 특성이 심혈관 질환을 예측하는 데 있어서 효과적인 지표로 활발히 활용되기때문에 결측치 대체 후 모델링 시도</br>

**scikit-learn**의 **iterativeimputer**를 사용하여 결측치 대체</br>
다른 특성과의 관계를 학습 후 결측치를 회귀 예측하는 방법</br>
[scikit-learn 결측치 처리 기법 공식문서](https://scikit-learn.org/stable/modules/impute.html#iterative-imputer)

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer

imputer = IterativeImputer(max_iter=10, random_state=77)
imputed_missingvalues = imputer.fit_transform(df)
imputed = pd.DataFrame(imputed_missingvalues)
imputed
```

```python
imputed.columns = ['성별코드', '연령대', '신장(5Cm단위)', '허리둘레', '수축기 혈압', '이완기 혈압', '식전혈당(공복혈당)', '총 콜레스테롤', '중성지방', 
                   'HDL 콜레스테롤','LDL 콜레스테롤', '흡연상태', '음주여부', '허리신장비율']
df = imputed
df
```

### 결측치 특성 제거 후 최종점수 특성 생성

```python
# 최종점수 컬럼 생성 
df_deleted['최종점수'] = 0

# # 남성이며 45세 이상 또는 여성이며 55세 이상인 경우 +1 점
df_deleted.loc[(df_deleted['성별코드'] == 1) & (df_deleted['연령대'] >= 45), '최종점수'] += 1
df_deleted.loc[(df_deleted['성별코드'] == 2) & (df_deleted['연령대'] >= 55), '최종점수'] += 1

# # 저혈압 +1점, 고혈압 1기 +1점, 2기 +2점
df_deleted.loc[(df_deleted['수축기 혈압'] > 140) & (df_deleted['수축기 혈압'] <= 159) & (df_deleted['이완기 혈압'] > 90) & (df_deleted['이완기 혈압'] <= 99), '최종점수'] += 1
df_deleted.loc[(df_deleted['수축기 혈압'] >= 160) & (df_deleted['이완기 혈압'] >= 100), '최종점수'] += 2
df_deleted.loc[(df_deleted['수축기 혈압'] <= 90) & (df_deleted['이완기 혈압'] <= 60), '최종점수'] += 1

# # 공복혈당 전 단계 +1점, 당뇨단계 +2점
df_deleted.loc[(df_deleted['식전혈당(공복혈당)'] > 100) & (df_deleted['식전혈당(공복혈당)'] <= 125), '최종점수'] += 1
df_deleted.loc[(df_deleted['식전혈당(공복혈당)'] >= 126), '최종점수'] += 2

# # 흡연자일 경우 +1점
df_deleted.loc[df_deleted['흡연상태'] == 1, '최종점수'] += 1

# # 음주자일 경우 +1점
df_deleted.loc[df_deleted['음주여부'] == 1, '최종점수'] += 1

# # 매우 마름, 과체중일 경우 +1점, 위험일 경우 +2점, 고위험일 경우 +3점
# # 남성
df_deleted.loc[(df_deleted['허리신장비율'] <= 0.34) & (df_deleted['성별코드'] == 1), '최종점수'] += 1
df_deleted.loc[(df_deleted['허리신장비율'] > 0.52) & (df_deleted['허리신장비율'] <= 0.57) & (df_deleted['성별코드'] == 1), '최종점수'] += 1
df_deleted.loc[(df_deleted['허리신장비율'] > 0.57) & (df_deleted['허리신장비율'] <= 0.62) & (df_deleted['성별코드'] == 1), '최종점수'] += 2
df_deleted.loc[(df_deleted['허리신장비율'] > 0.62) & (df_deleted['성별코드'] == 1), '최종점수'] += 3
# # 여성
df_deleted.loc[(df_deleted['허리신장비율'] <= 0.34) & (df_deleted['성별코드'] == 2), '최종점수'] += 1
df_deleted.loc[(df_deleted['허리신장비율'] > 0.48) & (df_deleted['허리신장비율'] <= 0.53) & (df_deleted['성별코드'] == 2), '최종점수'] += 1
df_deleted.loc[(df_deleted['허리신장비율'] > 0.53) & (df_deleted['허리신장비율'] <= 0.57) & (df_deleted['성별코드'] == 2), '최종점수'] += 2
df_deleted.loc[(df_deleted['허리신장비율'] > 0.57) & (df_deleted['성별코드'] == 2), '최종점수'] += 3

minimum_score_deleted = df_deleted['최종점수'].min()
maximum_score_deleted = df_deleted['최종점수'].max()
print(f'최종점수 최소값 : {minimum_score_deleted}')
print(f'최종점수 최대값 : {maximum_score_deleted}')

# 최종점수 특성 생성된 것 확인
df_deleted
```

### 결측치 특성 대체 후 최종점수 특성 생성

```python
# 최종점수 컬럼 생성 
df['최종점수'] = 0

# 남성이며 45세 이상 또는 여성이며 55세 이상인 경우 +1 점
df.loc[(df['성별코드'] == 1) & (df['연령대'] >= 45), '최종점수'] += 1
df.loc[(df['성별코드'] == 2) & (df['연령대'] >= 55), '최종점수'] += 1

# 고혈압 1기 +1점, 2기 +2점
df.loc[(df['수축기 혈압'] > 140) & (df['수축기 혈압'] <= 159) & (df['이완기 혈압'] > 90) & (df['이완기 혈압'] <= 99), '최종점수'] += 1
df.loc[(df['수축기 혈압'] >= 160) & (df['이완기 혈압'] >= 100), '최종점수'] += 2
df.loc[(df['수축기 혈압'] <= 90) & (df['이완기 혈압'] <= 60), '최종점수'] += 1

# 공복혈당 전 단계 +1점, 당뇨단계 +2점
df.loc[(df['식전혈당(공복혈당)'] > 100) & (df['식전혈당(공복혈당)'] <= 125), '최종점수'] += 1
df.loc[(df['식전혈당(공복혈당)'] >= 126), '최종점수'] += 2

# 총 콜레스테롤
df.loc[(df['총 콜레스테롤'] >= 140) & (df['총 콜레스테롤'] <= 239), '최종점수'] += 1
df.loc[(df['총 콜레스테롤'] >= 240), '최종점수'] += 2
        
# 중성지방
df.loc[(df['중성지방'] >= 150) & (df['중성지방'] <= 199), '최종점수'] += 1
df.loc[(df['중성지방'] >= 200) & (df['중성지방'] <= 499), '최종점수'] += 2
df.loc[(df['중성지방'] >= 500), '최종점수'] += 3
        
# LDL 콜레스테롤(양성요인, 높을수록 몸에 안 좋은 것)
df.loc[(df['LDL 콜레스테롤'] >= 130) & (df['LDL 콜레스테롤'] <= 159), '최종점수'] += 1
df.loc[(df['LDL 콜레스테롤'] >= 160) & (df['LDL 콜레스테롤'] <= 189), '최종점수'] += 2
df.loc[(df['LDL 콜레스테롤'] >= 190), '최종점수'] += 3
        
# HDL 콜레스테롤(음성요인, 높을수록 몸에 좋은 것)
df.loc[(df['HDL 콜레스테롤'] < 40), '최종점수'] += 1
df.loc[(df['HDL 콜레스테롤'] >= 60), '최종점수'] -= 1

# 흡연자일 경우 +1점
df.loc[df['흡연상태'] == 1, '최종점수'] += 1

# 음주자일 경우 +1점
df.loc[df['음주여부'] == 1, '최종점수'] += 1

# 매우 마름, 과체중일 경우 +1점, 위험일 경우 +2점, 고위험일 경우 +3점
# 남성
df.loc[(df['허리신장비율'] <= 0.34) & (df['성별코드'] == 1), '최종점수'] += 1
df.loc[(df['허리신장비율'] > 0.52) & (df['허리신장비율'] <= 0.57) & (df['성별코드'] == 1), '최종점수'] += 1
df.loc[(df['허리신장비율'] > 0.57) & (df['허리신장비율'] <= 0.62) & (df['성별코드'] == 1), '최종점수'] += 2
df.loc[(df['허리신장비율'] > 0.62) & (df['성별코드'] == 1), '최종점수'] += 3
# 여성
df.loc[(df['허리신장비율'] <= 0.34) & (df['성별코드'] == 2), '최종점수'] += 1
df.loc[(df['허리신장비율'] > 0.48) & (df['허리신장비율'] <= 0.53) & (df['성별코드'] == 2), '최종점수'] += 1
df.loc[(df['허리신장비율'] > 0.53) & (df['허리신장비율'] <= 0.57) & (df['성별코드'] == 2), '최종점수'] += 2
df.loc[(df['허리신장비율'] > 0.57) & (df['성별코드'] == 2), '최종점수'] += 3

minimum_score = df['최종점수'].min()
maximum_score = df['최종점수'].max()
print(f'최종점수 최소값 : {minimum_score}')
print(f'최종점수 최대값 : {maximum_score}')

# 최종점수 특성 생성된 것 확인
df
```
